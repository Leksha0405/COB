```python
import requests
import pandas as pd
import os

url='https://api.binance.com/api/v3/ticker/24hr'

response= requests.get(url)
if response.status_code == 200:
    data = response.json()
    df= pd.DataFrame(data)
    
    documents_path= os.path.expanduser("~/Documents")
    csv_filename = os.path.join(documents_path, 'binance_data.csv')
    df.to_csv(csv_filename, index=False)
    print(f"Data has been saved to {csv_filename}")
else:
    print(f"Error: Unable to fetch data from Binance API. Status code: {response.status_code}")
```

```python
import pandas as pd
import os 

def remove_outliers(data, columns, threshold=3):
    z_scores = (data[columns] - data[columns].mean()) / data[columns].std()
    data_no_outliers = data[(z_scores.abs() < threshold).all(axis=1)]
    return data_no_outliers

numeric_column = ['release_year']
data_no_outliers = remove_outliers(data, numeric_column)
documents_path = os.path.expanduser("~/Documents")
outliers_removed_fileName = os.path.join(documents_path, 'data_no_outlier.csv')
data_no_outlier.to_csv(outliers_removed_fileName, index=False)
print(f"Data without has been saved to {outliers_removed_fileName}")
```
